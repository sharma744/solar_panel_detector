```python
import cv2
from tensorflow import keras;
from keras.utils import image_dataset_from_directory
from keras.preprocessing import image
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
arr=image_dataset_from_directory("/Users/aayushsharma/Downloads/Faulty_solar_panel",
    labels="inferred",
    label_mode="int",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(256, 256)
    )
```

```python
arr=arr.map(process)
```

```python
def process(images, labels):

    images = tf.cast(images, tf.float32) / 255.0

    return images, labels
```

```python
from keras.applications import VGG16
```

```python
model1 = VGG16(
    weights="imagenet",
    input_shape=(256,256,3),
    include_top=False
)

# STEP 1: unfreeze all
model1.trainable =False

# STEP 2: freeze all layers first
for layer in model1.layers:
    layer.trainable = False

# STEP 3: unfreeze last 10 layers
for layer in model1.layers[-10:]:
    layer.trainable = True
```

```python
for images,labels in arr:
    print(labels)
```

```python
from keras.layers import Dense,BatchNormalization,Dropout,Flatten,GlobalAveragePooling2D
from keras.models import Sequential
```

```python
model=Sequential()
model.add(model1)
model.add(GlobalAveragePooling2D())
model.add(Dense(128,activation="relu"))
model.add(BatchNormalization())
model.add(Dropout(0.2))
# model.add(Dense(32,activation="relu"))
# model.add(BatchNormalization())
# model.add(Dropout(0.3))
model.add(Dense(6,activation="softmax"))
```

```python
model.summary()
```

```python
from keras.optimizers import Adam
adam=Adam(learning_rate=0.001)
model.compile(loss="sparse_categorical_crossentropy",optimizer=adam,metrics=["accuracy"])
```

```python
model.fit(arr,epochs=25)
```

```python
model.save("solear.keras")
```
